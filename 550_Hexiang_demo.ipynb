```python
cast=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/550/title.principals.tsv',sep='\t',header=0)
cast = cast.loc[cast['characters'] != '\\N']
cast['characters']=cast['characters'].str.replace(r'\[\"', '') 
cast['characters']=cast['characters'].str.replace(r'\"\]', '') 
cast['characters']=cast['characters'].str.strip()
cast.head()
```
